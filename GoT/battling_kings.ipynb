```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
csv_file = "Resources/got.csv"

got_df = pd.read_csv(csv_file)

got_df.head()
```

```python
# Get attacker and defender data
atk_def_data = got_df.loc[:, ["attacker_king", "defender_king", "battle_number"]]
atk_def_data.head()
```

```python
# grab attack and defend data for each king using groupby
atk_data = atk_def_data.groupby("attacker_king").count()["battle_number"]
atk_data
```

```python
# create bar chart graphing the number of times each king was the attacker
atk_data.plot(kind="bar", facecolor="red")

# format bar chart
plt.title("Kings of GOT")
plt.xlabel("King")
plt.ylabel("Number of Battles")

plt.show()
plt.savefig("attack_chart.png")
plt.tight_layout()
```

```python
# create bar chart graphing the number of times each king was the defender
def_data = atk_def_data.groupby("defender_king").count()["battle_number"]

def_data.plot(kind="bar", facecolor="blue")

# format bar chart
plt.title("Kings of GOT")
plt.xlabel("King")
plt.ylabel("Number of Battles Defended")

plt.show()
plt.savefig("defender_chart.png")
plt.tight_layout()
```
